```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GCNConv, global_mean_pool
```

```python
connectivity_cols = [col for col in df_reduced.columns if col not in ['participant_id', 'ADHD_Outcome', 'Sex_F']]
```

```python
# ----- Helper Function to Convert a Connectivity Vector to a Graph -----
def create_graph_from_row(row, connectivity_cols, threshold=0):
    """
    Converts a row from df_reduced into a torch_geometric Data object.
    The connectivity vector is converted into a connectivity matrix.
    
    If the vector length is a perfect square, it is assumed to be a full connectivity matrix.
    If not, the function attempts to see if the vector length matches that of the
    upper-triangular portion of a symmetric matrix and reconstructs the full matrix.
    """
    # Extract connectivity values and convert to float
    conn_vals = row[connectivity_cols].values.astype(float)
    n_elements = len(conn_vals)
    
    # First, check if it's a full connectivity matrix (perfect square length)
    sqrt_val = np.sqrt(n_elements)
    if sqrt_val.is_integer():
        n_nodes = int(sqrt_val)
        conn_matrix = conn_vals.reshape(n_nodes, n_nodes)
    else:
        # Check if the vector length matches the upper triangular part (without the diagonal)
        # For a symmetric matrix without self-connections, the number of elements should be n*(n-1)/2.
        n_nodes_float = (1 + np.sqrt(1 + 8 * n_elements)) / 2
        if n_nodes_float.is_integer():
            n_nodes = int(n_nodes_float)
            # Initialize an empty matrix
            conn_matrix = np.zeros((n_nodes, n_nodes))
            # Get the indices for the upper triangle (excluding the diagonal)
            triu_indices = np.triu_indices(n_nodes, k=1)
            if len(conn_vals) != len(triu_indices[0]):
                raise ValueError(f"Unexpected mismatch in triangular indices for participant {row['participant_id']}.")
            # Fill the upper triangle with the connectivity values
            conn_matrix[triu_indices] = conn_vals
            # Mirror to the lower triangle to obtain a symmetric matrix
            conn_matrix = conn_matrix + conn_matrix.T
        else:
            raise ValueError(f"Participant {row['participant_id']} connectivity vector length is not a perfect square nor matches a triangular number.")
    
    # Convert the connectivity matrix to a torch tensor
    matrix = torch.tensor(conn_matrix, dtype=torch.float)
    
    # Create edge_index based on threshold (you may adjust threshold if needed)
    edge_index = (matrix > threshold).nonzero(as_tuple=False).t().contiguous()
    # Get edge attributes (weights) and add a feature dimension
    edge_attr = matrix[edge_index[0], edge_index[1]].unsqueeze(1)
    
    # Use an identity matrix as node features (adjust size if necessary)
    x = torch.eye(n_nodes, dtype=torch.float)
    
    # Create label tensor. For binary classification with BCEWithLogitsLoss, ensure label is 0 or 1.
    label = torch.tensor([row['ADHD_Outcome']], dtype=torch.float)
    
    # Create and return a Data object (optionally, store participant_id)
    data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr, y=label)
    data.participant_id = row['participant_id']
    return data
```

```python
# ----- Build a List of Graph Data Objects -----
data_list = []
# IMPORTANT: Ensure connectivity_cols includes only your connectivity matrix columns.
for idx, row in df_reduced.iterrows():
    try:
        graph_data = create_graph_from_row(row, connectivity_cols, threshold=0)
        data_list.append(graph_data)
    except Exception as e:
        print(f"Skipping participant {row['participant_id']} due to error: {e}")
```

```python
#----- Split Data into Training and Testing Sets 80-20 spliy -----
if len(data_list) == 0:
    raise ValueError("No valid graph examples were generated. Check your connectivity_cols and data formatting.")

split_index = int(0.8 * len(data_list))
train_data = data_list[:split_index]
test_data = data_list[split_index:]
print("Number of training examples:", len(train_data))
print("Number of testing examples:", len(test_data))
```

```python
#----- Create DataLoaders for Batching -----
train_loader = DataLoader(train_data, batch_size=8, shuffle=True)
test_loader = DataLoader(test_data, batch_size=8, shuffle=False)
```

```python
#----- Graph Neural Network Model -----
class GCN(nn.Module):
    def __init__(self, in_channels, hidden_channels, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        # Final linear layer outputs one logit per graph (for binary classification)
        self.lin = nn.Linear(hidden_channels, num_classes)
    
    def forward(self, data):
        x, edge_index, edge_attr, batch = data.x, data.edge_index, data.edge_attr, data.batch
        x = F.relu(self.conv1(x, edge_index, edge_attr))
        x = F.relu(self.conv2(x, edge_index, edge_attr))
        # obtain a graph-level representation
        x = global_mean_pool(x, batch)
        return self.lin(x).view(-1)  # Output shape: (batch_size,)
```

```python
# input channels from node features (here, the identity matrix size)
sample_n_nodes = train_data[0].x.size(0)
in_channels = train_data[0].x.size(1)  #equals sample_n_nodes
```

```python
# model, optimizer, loss function
model = GCN(in_channels=in_channels, hidden_channels=16, num_classes=1)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.BCEWithLogitsLoss()
```

```python
#---- Training Loop -----
num_epochs = 100
model.train()
for epoch in range(num_epochs):
    total_loss = 0
    for batch in train_loader:
        optimizer.zero_grad()
        out = model(batch)
        loss = criterion(out, batch.y.view(-1))
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss / len(train_loader):.4f}")
```

```python
#---- Evaluation on Test Data -----
model.eval()
correct = 0
total = 0
for batch in test_loader:
    out = model(batch)
    preds = (torch.sigmoid(out) > 0.5).float()
    correct += (preds == batch.y.view(-1)).sum().item()
    total += batch.num_graphs  # Each graph corresponds to one example
print(f"Test Accuracy: {correct / total:.4f}")
```
